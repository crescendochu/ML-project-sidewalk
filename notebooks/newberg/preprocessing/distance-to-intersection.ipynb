```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
%matplotlib inline
import seaborn as sns; sns.set_theme(color_codes=True)
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
```

```python
city = 'newberg'
```

```python
city_crs = {'seattle': 2285, oradell: 3424, 'chicago': 3435, 'newberg':2269}
```

```python
#read in intersections data
intersections = gpd.read_file('../../../data/{}/osm/osm-intersections/osm-intersections-cleaned/osm-intersections-cleaned.shp'.format(city))
intersections
```

```python
#read in labels
labels = gpd.read_file('../../../data/{}/processed-labels/labels_raw_{}/labels_raw_{}.shp'.format(city, city, city))
labels
```

```python
#select the curb ramps and non-curb ramps
curb = labels.loc[(labels['label_type']=='CurbRamp')]
no_curb = labels.loc[(labels['label_type']=='NoCurbRamp')]
#concatenate the two dataframes
all_curbs = pd.concat([curb, no_curb])
all_curbs
```

```python
#change crs to 3424
intersections = intersections.to_crs(epsg= city_crs[city])
all_curbs = all_curbs.to_crs(epsg= city_crs[city])
```

```python
#spatial join curb ramps to intersections
curb_to_intersection = all_curbs.sjoin_nearest(intersections, distance_col='distance')
curb_to_intersection
```

```python
#select only label_id, distance
curb_to_intersection = curb_to_intersection[['label_id','distance']]
```

```python
#rename 
curb_to_intersection = curb_to_intersection.rename(columns={'distance':'distance_to_intersection'})
curb_to_intersection.head()
```

```python
#merge labels and intersection_d
df = labels[['label_id']].merge(curb_to_intersection, on='label_id', how='left')
#fill in missing values with 0
df = df.fillna(0)
df
```

```python
df.to_csv('../../../data/{}/features/distance_to_intersection.csv'.format(city), index=False)
```
