```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
%matplotlib inline
import seaborn as sns; sns.set_theme(color_codes=True)
pd.set_option('display.max_columns', None)
import os
```

```python
city = 'newberg'
```

```python
#read in streets data
streets = gpd.read_file('../../../../data/{}/source/streets.json'.format(city))
```

```python
#select only street_edge_id,osm_id,geometry
streets = streets[['street_edge_id','osm_id','geometry']]
streets
streets.sort_values(by=['street_edge_id'])
```

```python
#streets.explore(tiles='cartodbpositron')
```

```python
#read in osm_id to waytype mapping
mapping = pd.read_csv('../../../../data/{}/source/streets-with-osm-way-info-{}.csv'.format(city,city))
```

```python
mapping
```

```python
#inner join the two dataframes on street_edge_id
streets = pd.merge(streets, mapping, on='street_edge_id', how='inner')
streets
```

```python
#streets.explore(tiles='cartodbpositron', column='way_type')
```

```python
#change the order of columns geometry to the end
streets = streets[['street_edge_id','osm_id','way_type','geometry']]
streets
```

```python
file_path = '../../../../data/{}/osm/osm-roads/streets-with-osm-way-type.shp'.format(city)
directory_path = os.path.dirname(file_path)
if not os.path.exists(directory_path):
    os.makedirs(directory_path)
```

```python
#save the streets data as shapefile 
streets.to_file(file_path)
```
