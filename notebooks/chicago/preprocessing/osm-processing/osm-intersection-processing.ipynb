```python
import geopandas as gpd
import pandas as pd
import os
```

```python
city = 'chicago'
```

```python
intersections_geo = gpd.read_file('../../../../data/'+ city + '/osm/osm-intersections/intersections-with-correct-geo/intersections-with-correct-geo.shp')
intersections_waytype = gpd.read_file('../../../../data/' + city + '/osm/osm-intersections/intersections-with-way-type/intersections-with-way-type.shp')
```

```python
#spatial join two dataframes
df = gpd.sjoin(intersections_geo, intersections_waytype, how='inner')
df
```

```python
#keep highways osm_id and geometry from df
intersections = df[['way_type','osm_id', 'geometry']]
intersections
```

```python
file_path = '../../../../data/'+ city +'/osm/osm-intersections/osm-intersections-cleaned/osm-intersections-cleaned.shp'
directory_path = os.path.dirname(file_path)
if not os.path.exists(directory_path):
    os.makedirs(directory_path)
intersections.to_file(file_path)
```
