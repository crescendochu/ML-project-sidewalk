```python
import geopandas as gpd
import pandas as pd
```

```python
intersections_geo = gpd.read_file('../../../../data/seattle/osm/osm-intersections/intersections-with-correct-geo/intersections-with-correct-geo.shp')
intersections_waytype = gpd.read_file('../../../../data/seattle/osm/osm-intersections/intersections-with-way-type/intersections-with-way-type.shp')
```

```python
#spatial join two dataframes
df = gpd.sjoin(intersections_geo, intersections_waytype, how='inner')
df
```

```python
#keep highways osm_id and geometry from df
intersections = df[['highway','osmid', 'geometry']]
intersections
```

```python
#save to shapefile
intersections.to_file('data/osm/osm-intersections/osm-intersections-cleaned/osm-intersections-cleaned.shp')
```
