```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
%matplotlib inline
import seaborn as sns; sns.set_theme(color_codes=True)
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors
import warnings
warnings.filterwarnings('ignore')
from scipy import stats
import math
import folium
```

```python
# read label.csv
df = pd.read_csv('data/labels.csv')
```

```python
label_colors ={'Crosswalk':'#FABF1C',
               'CurbRamp':'#90C31F',
               'NoCurbRamp':'#E679B6',
               'NoSidewalk':'#BE87D8',
               'Obstacle' :'#78B0EA',
               'Occlusion':'#B3B3B3',
               'Other':'#B3B3B3',
               'Signal':'#63C0AB',
               'SurfaceProblem':'#F68D3E'}
```

```python
sns.palplot(sns.color_palette(label_colors.values()))
```

```python
# set white style background
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(9,6))
ax=sns.countplot(x="label_type", data=df, palette =label_colors)
for container in ax.containers:
    ax.bar_label(container)

# add x y label
plt.xlabel('', fontsize=14)
plt.ylabel('', fontsize=14)
#change x label ticks size
plt.xticks(fontsize=12)
#change x label font weight
plt.xticks(fontweight='bold')
#change x label ticks direction
plt.xticks(rotation=45)
#save figure to png
plt.savefig('data/df_train_label_count.png', dpi=300, bbox_inches='tight')
```
