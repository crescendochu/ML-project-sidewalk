```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
%matplotlib inline
import seaborn as sns; sns.set_theme(color_codes=True)
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors
import warnings
warnings.filterwarnings('ignore')
from scipy import stats
import math
```

```python
df = pd.read_csv('data/seattle_labels.csv')
```

```python
gt = pd.read_csv('data/test_set_seattle.csv')
```

```python
#select only label_id and verified
gt = gt[['label_id', 'verified']]
```

```python
#change verified column name to correct
gt = gt.rename(columns={'verified': 'correct'})
```

```python
fp = pd.read_csv('data/features/false_positives.csv')
```

```python
#select false positives = 1
fp = fp[fp['false_positive'] == 1]
```

```python
fp
```

```python
#add new column with correct = 0
fp['correct'] = 0
#drop false positive column
fp = fp.drop(columns=['false_positive'])
```

```python
gt
```

```python
fp
```

```python
#cocatenate gt and fp
dt = pd.concat([gt, fp])
dt
```

```python
#group by label_id and max
dt = dt.groupby('label_id').max()
#reset index
dt = dt.reset_index()
```
