```python
# Define a function to calculate the mean HSV value of an image
def mean_hsv(img_path):
    img = cv2.imread(img_path)
    img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    mean_h, mean_s, mean_v = np.mean(img_hsv, axis=(0, 1))
    return mean_h, mean_s, mean_v

# Set thresholds for matching (depends on how exact the match should be)
THRESHOLD = 1  # Example threshold, adjust as necessary

# Get the mean HSV value for the reference image
ref_h, ref_s, ref_v = mean_hsv('26.jpg')
print(ref_h, ref_s, ref_v)

# Source and destination directories
src_dir = 'C:\\Users\\Zhihan Zhang\\Downloads\\Test'  # Update with your source folder
dst_dir = 'C:\\Users\\Zhihan Zhang\\Downloads\\Test\\Trash'  # Update with your desired destination folder
if not os.path.exists(dst_dir):
    os.makedirs(dst_dir)

# Iterate through each image in the source directory
for img_name in os.listdir(src_dir):
    print(img_name)
    img_path = os.path.join(src_dir, img_name)
    if not img_path.endswith('.jpg'):  # skip non-jpg files
        continue

    # Get mean HSV for current image
    h, s, v = mean_hsv(img_path)
    print(h, s, v)

    # Check if the HSV values are within the threshold of the reference image's values
    if (abs(h - ref_h) <= THRESHOLD) and (abs(s - ref_s) <= THRESHOLD) and (abs(v - ref_v) <= THRESHOLD):
        # Move the image to the new folder
        shutil.move(img_path, os.path.join(dst_dir, img_name))

print('Done!')
```
